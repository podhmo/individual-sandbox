```python
import numpy as np
import pandas as pd
import numpy.random as r
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline
```

```python
N = 10000
df = pd.DataFrame(r.ranf(size=(N,1)), columns=["n"])
df.plot.hist(bins=10)
```

```python
df = pd.DataFrame(r.ranf(size=(N,1)),columns=["n"])
p = 0.6
df["p"] = df["n"] > p
df["p"].plot.hist(bins=2)
```

```python
df = pd.DataFrame(r.ranf(size=(N, 3)))
df2 = df[:] > p
df3 = pd.DataFrame(df2.apply(lambda xs : "".join([str(int(x)) for x in xs]), axis=1), columns=["r"])
df4 = (df3.groupby("r")["r"].count() / N)
df4.plot.bar()
df4.sum()
```

```python
df = pd.DataFrame(r.ranf(size=(N, 10)))
df2 = df[:] > p
df3 = pd.DataFrame(df2.apply(lambda xs : np.sum(xs), axis=1), columns=["r"])
df4 = (df3.groupby("r")["r"].count() / N)
df4.plot.bar()
df4.sum()
```

```python
df = pd.DataFrame(r.ranf(size=(N, 30)))
df2 = df[:] > p
df3 = pd.DataFrame(df2.apply(np.sum, axis=1), columns=["r"])
df4 = (df3.groupby("r")["r"].count() / N)
df4.plot.bar()
df4.sum()
```
